```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2 * np.pi, 0.01)
x = np.cos(t)
y = np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.grid()
ax.set_aspect(1)

plt.show()
```
